# 12.2 计算机视觉      
## 12.2.1 对比度归一化     
对比度指的是图像中亮相素与暗像素之间差异的大小，一般通过图像中像素的标准差来量化，如对应有红绿蓝三个channel的图片，对比度可以表示如下：    
&emsp;$ \sqrt{\frac{1}{3rc}\sum\limits^r_{i=1}\sum\limits^c_{j=1}\sum\limits^3_{k=1}(X_{i,j,k}-\bar X)^2}$     
* **全局对比度归一化**： 通过从每个图像中减去平均值，然后重新缩放使像素上的标准差等于某个常数s,防止输入的图像数据集具有变化的对比度。     
* **局部对比度归一化**： 通过减去邻近像素的平均值并除以邻近像素的标准差，其结果是会丢弃所有强度相同的区域，是模型只关注边缘。    

## 12.2.2 数据集增强      
增加训练集的额外副本（可以对原始数据做一些不改变类别的变换来完成）来增加训练集大小，并提高泛化能力。

## 12.4.3 高维输出      
许多自然语言应用中，通常像模型生产词作为输出的基本单位，涉及到大词汇量表，由于词汇量大，在词的选择上表示输出分布的计算成本可能非常高。     
### 12.4.3.1 使用短列表      
有一种方法是将词汇表分为最常见词汇的短列表和较稀有词汇的尾列表，为此神经网络需要预测词位于尾列表的概率。      
明显的缺点是：神级网络语言模型的潜在泛化趋势最终会仅限于使用最常用的词。
