**参考的实践设计流程**：    
1. 确定目标：定义损失函数。    
2. 建立端到端的工作流程，包括设计合适的性能度量。     
3. 搭建系统：确定性能瓶颈，检查哪个部分低于预期，是否过拟合、欠拟合，找到数据或软件缺陷。    
4. 反复进行增量式改动，如收集新数据、调整超参数和改进算法。

## 11.4.3 随机搜索      
为每个超参数定义一个边缘分布，不需要像网格搜索那样离散化超参数的值。当有几个超参数对性能没有显著影响时，随机搜索比网格搜索指数级的高效。     
# 11.5 调试策略
一些重要的调试策略：     
1. 可视化计算中模型的行为
2. 可视化最严重的错误
3. 根据训练和测试误差检测软件
4. 拟合极小的数据集
5. 比较反向传播导数和数值导数
6. 监控激活函数和梯度的直方图
